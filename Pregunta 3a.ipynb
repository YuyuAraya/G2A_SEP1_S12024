```python
import numpy as np

# Para el sistema a
potencia_va = 1200000000000  # Potencia en VA
factor_potencia = 0.9  # Factor de potencia

potencia_w = potencia_va * factor_potencia  # Potencia en W

print("Potencia en watts:", potencia_w)

P_indicada = potencia_w  # Potencia indicada en watts
P_min = P_indicada * 0.5  # Potencia mínima (50% por debajo de la potencia indicada)
P_max = P_indicada * 1.5  # Potencia máxima (50% por encima de la potencia indicada)

print(potencia_w, P_min, P_max)

num_cargas = 100  # Número de valores de carga a generar
cargas = np.linspace(P_min, P_max, num_cargas)
k = 2  # Valor constante
voltajes = k * np.sqrt(cargas)
import matplotlib.pyplot as plt

plt.plot(cargas / 1000000, voltajes / 1000)
plt.xlabel('Carga (MW)')
plt.ylabel('Voltaje (kV)')
plt.title('Voltaje vs Carga')
plt.grid(True)
plt.show()
```
